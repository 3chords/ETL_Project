```python
# import the dependency
import pandas as pd
import numpy as np
```

```python
# files for our analysis
file1 = "cre_stats_data.csv"
file2 = 'labor_data.csv'
```

```python
# load the labor file
labor_data = pd.read_csv(file2)
labor_data.head()
```

# add the MSA, MSA name to the local data ( use as primary keys to joint later)

```python
# to get the city and state data
city_state = cre_stats['Geography Name'].to_list()
```

```python
# split the city and state data to city and state
# list of cities
import re
city_list= []

for i in city_state:
    k = re.split(" - ", i)[0].split('-')[0]
    city_list.append(k)
   
```

```python
# list of states
state_list = []
for i in city_state:
    k = i.split(' ')[-1]
    state_list.append(k)
```

```python
# get msa city and state data from MSA NAME and create list of  citiea and list of states

MSA_name = labor_data['MSA Name']
MSA = labor_data['MSA']
MSA_city =[]
MSA_state = []
for i in MSA_name:
    MSA_city.append(i.split(",")[0])
    MSA_state.append(i.split(",")[1])
    
    
```

```python
# MSA city list from the labor data

MSA_city_list = []
for i in MSA_city:
    MSA_city_list.append(i.split("-"))
```

```python
# MSA state list from the labor data

MSA_state_list =[]


for i in MSA_state:
    MSA_state_list.append(i.split(' ')[1].split('-'))

```

```python
# assign the lenth of labor data to a variable
ml = len(MSA_state_list)
```

```python
# assign the lenth of cre state data to a variable
cl = len(city_list)
```

One of the challenges of uniting the two files was that the CRE Stats file had the time series data in rows but the Labor file had the time series data in columns and we wanted the final file to have the same format as the CRE Stats file. After much discussion, the team used Excel to investigate how to go about converting the Labor Data file into the CRE Stats format. We developed an intermediary list of dates and columns that corresponded to the data columns in the Labor data file. This intermediary list allowed us to build a nested vlookup function that simultaneously searched on both the MSA code and the year of data to pull the proper data in. This gave us a clean file to work with.
